```python
import pandas as pd
```

```python
df = pd.read_csv('FINAL_DF.csv')
df
```

```python
conf = pd.read_csv('CONF.CSV').drop_duplicates()
conf
```

```python
def check_names(dataset1, dataset2):
    bad_names = []
    for w in dataset1["Team"]:
        if w not in dataset2["School"].values and w not in dataset2["Opponent"].values and w not in bad_names:
            bad_names.append(w)
    print(len(bad_names))
    print(bad_names)
```

```python
school_merged = pd.merge(left=df, right=conf, how='left', left_on=['School'], right_on=['Team'])

opp_merged = pd.merge(left=df, right=conf, how='left', left_on=['Opponent'], right_on=['Team'])
opp_merged = opp_merged.rename(columns={
    "Team": "OppTeam",
    "Nickname": "OppNickname",
    "City": "OppCity",
    "State": "OppState",
    "Enrollment": "OppEnrollment",
    "Conference": "OppConference"
})[["School", "Opponent", "Date", "OppNickname", "OppCity", "OppState", "OppEnrollment", "OppConference"]]
merge_on = ['School', 'Opponent', 'Date']

merged_conferences = pd.merge(left=school_merged, right=opp_merged, how='inner', left_on=merge_on, right_on=merge_on).drop_duplicates().dropna()
```

```python
merged_conferences.to_csv('FINAL_DF_CONFERENCES.csv', index=False)
```

```python
import pandas as pd
```

Read in Stats

```python
stats = pd.read_csv('GameStats.csv')
stats.rename(columns={'G#': 'GameNumber',
                      'Unnamed: 4':'location',
                      'Unnamed: 6': 'Outcome'}, inplace=True)
stats.head()
```

Read in Results

```python
results = pd.read_csv('GameResults.csv')
results.nunique()
```

Merge Datasets

```python
merged_home = pd.merge(results, stats, left_on=['Home Team', 'Visitor Team', 'Date'], right_on=['School', 'Opponent', 'Date'], how='inner')
```

```python
merged_away = pd.merge(results, stats, left_on=['Visitor Team', 'Home Team', 'Date'], right_on=['School', 'Opponent', 'Date'], how='inner')
```
